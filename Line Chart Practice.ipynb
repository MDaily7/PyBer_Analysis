```python
# import matplotlib
import matplotlib.pyplot as plt
```

```python
# Create lists for axes data
savings = [0, 100, 150, 200, 300, 500, 700, 650, 800, 850, 900, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Months')
plt.ylabel('Savings ($)')
plt.title('Savings Results')
plt.show()
```
