```python
# import matplotlib
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# plot the previous lists
plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# method 2 to create a plot with ax.plot
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')
#Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')
# set the y limit between 0 and 45
plt.ylim(0, 45)
# create a title
plt.title("PyBer Fare by Month")
# add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="Purple", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# use object oriented approach to make a more formatted figure
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker='D', linewidth=2, label='Boston')
# axes labels and title
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
# set limit to y axis
ax.set_ylim(0, 45)
# add grid lines
ax.grid()
# add the legend
ax.legend()
```

```python
# create vertical bar chart
plt.bar(x_axis, y_axis, label='Boston', color='Black')
# Create labels
plt.xlabel('Date')
plt.ylabel('Fare ($)')
# title
plt.title('PyBer Fare by Month')
# add a legend
plt.legend()
```

```python
# create horizontal bar chart
plt.barh(x_axis, y_axis, label='Boston', color='Magenta')
# invert y axis
plt.gca().invert_yaxis()
# Create labels
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# title
plt.title('PyBer Fare by Month')
# add a legend
plt.legend()
```

```python
# object oriented vertical bar chart
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="Black", label='New York')
ax.legend()
```

```python
# horizontal bar chart with object oriented approach
fig, ax = plt.subplots()
plt.gca().invert_yaxis()
ax.barh(x_axis, y_axis, color="Cyan", label='Chicago')
ax.legend()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
```

```python
# creating a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# creating a scatter plot method 2 of matlab methods
plt.scatter(y_axis, x_axis , c='red', label='Chicago', s=y_axis, marker='x')
# title and axes lables
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# add limit to x-axis data
plt.xlim(0, 45)
# invert y axis
plt.gca().invert_yaxis()
```

```python
# creating a bubble chart
# creating a scatter plot method 2 of matlab methods
plt.scatter(y_axis, x_axis , c='black', label='Chicago', s=y_axis)
# title and axes lables
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# add limit to x-axis data
plt.xlim(0, 45)
# invert y axis
plt.gca().invert_yaxis()
```

```python
# creating a list to use as larger sizes for bubble chart
y_axis_larger = []
# for loop to add to the list
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# creating a bubble chart
# creating a scatter plot method 2 of matlab methods
plt.scatter(y_axis, x_axis , c='black', label='Chicago', s=y_axis_larger)
# title and axes lables
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# add limit to x-axis data
plt.xlim(0, 45)
# invert y axis
plt.gca().invert_yaxis()
```

```python
# creating a bubble chart
# creating a scatter plot method 2 of matlab methods
plt.scatter(y_axis, x_axis , c='black', label='Chicago', s=[i * 3 for i in y_axis])
# title and axes lables
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# add limit to x-axis data
plt.xlim(0, 45)
# invert y axis
plt.gca().invert_yaxis()
```

```python
# creating the scatter plot using object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# creating the bubble plot using object oriented interface
fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c='skyblue', alpha=.2, linewidth=2, edgecolors='k', label='Boston')
# axes and titles
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
plt.gca().invert_yaxis()
ax.set_xlim(0, 50)
ax.legend()
```

```python
# Matlab method for creating a pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# adding bling to the pie chart
explode_values = (0, 0, 0, 0, 0, 0, .2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', colors=colors)
plt.show()
```

```python
# object oriented approach
explode_values = (0, 0, .2, 0, 0, 0, .2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values, shadow=True, startangle=98, colors=colors, counterclock=False)
plt.show()
```
